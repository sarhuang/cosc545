```python
def clean_crime_data(file_path):
    # Read a chunk of the data to inspect the file structure
    data_chunk = pd.read_csv(file_path, nrows=50)

    # Find the row number where the actual data starts
    start_of_data = None
    for index, row in data_chunk.iterrows():
        if row.isnull().sum() < len(row) - 2:  # Assuming data starts when there are more than 2 non-NA values
            start_of_data = index
            break

    # Read the data again, skipping rows until the start of data
    data_corrected = pd.read_csv(file_path, skiprows=start_of_data)

    # Use the first row as the header and drop it from the data
    data_cleaned = data_corrected.drop([0]).reset_index(drop=True)

    # Define new headers
    headers = ['Offense_Category', 'Subcategory', 'Total', 'Male', 'Female', 'White', 'Black', 'Hispanic']

    # Assign the new headers
    data_cleaned.columns = headers

    # Replace NaN values in 'Offense_Category' with the value from the previous row
    data_cleaned['Offense_Category'] = data_cleaned['Offense_Category'].fillna(method='ffill')

    # Remove any remaining header or note rows that don't contain data
    data_cleaned = data_cleaned.dropna(subset=['Subcategory'])

    return data_cleaned
problem_csv_files = []
clean_csv_files = []
for i in range(88):
    file_path = csv_files[i]
    try:
        dataframe = clean_crime_data(file_path)
        clean_csv_files.append(file_path)
        print(dataframe)
    except:
        problem_csv_files.append(file_path)
print("------------Below are all the csv files with different structure.... like additional column or not having proper structure--------")
print(problem_csv_files)
```

```python
drug_data = pd.read_csv('data/drug_data.csv')

# Convert the 'Date' column to datetime format
drug_data['Date'] = pd.to_datetime(drug_data['Date'], errors='coerce')

# Group data by year and sum the number of inmates
yearly_data = drug_data.groupby(drug_data['Date'].dt.year)['All inmates'].sum().reset_index()

# Plotting the trend of inmates over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_data, x='Date', y='All inmates')
plt.title('Trend of Inmates Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Inmates')
plt.grid(True)
plt.show()
```

```python
# Summing up male and female inmates
gender_data = drug_data[['Male', 'Female']].sum()

# Pie chart for gender distribution
plt.figure(figsize=(8, 8))
gender_data.plot.pie(autopct='%1.1f%%', startangle=140, colors=['#66b3ff','#ff9999'])
plt.title('Gender Distribution of Inmates')
plt.ylabel('')  # Hide the y-label
plt.show()
```

```python
# Summing up race data
race_columns = ['White', 'Other','Black','Asian', 'Hispanic','American Indian/Alaska Native']
race_data = drug_data[race_columns].sum()

# Pie chart for race distribution
plt.figure(figsize=(8, 8))
race_data.plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Race Distribution of Inmates')
plt.ylabel('')  # Hide the y-label
plt.show()
```

```python
unique_years = dates.dt.year.unique()

fig, ax = plt.subplots(figsize=(10, 6))

# Plotting as side-by-side bar graphs
width = 0.35  # Width of each bar
bar_positions_estimated = np.arange(len(unique_years))
bar_positions_actual = bar_positions_estimated + width

# Grouping the data by year and summing the values for each year
grouped_data = drug_data.groupby([dates.dt.year, 'Estimated or Actual'])['All inmates'].sum().unstack()

plt.bar(bar_positions_estimated, grouped_data['Estimated'], width, label='Estimated', color='blue')
plt.bar(bar_positions_actual, grouped_data['Actual'], width, label='Actual', color='orange')

plt.xlabel('Year')
plt.ylabel('Inmates')
plt.title('Estimated vs Actual Inmates Over Time')
plt.xticks(bar_positions_estimated + width / 2, unique_years)
plt.legend()

# Display the plot
plt.show()
```

```python
drug_all = np.array(drug_data[drug_data['Offense'] == 'Drug']['All inmates'])
possession_all = np.array(drug_data[drug_data['Offense'] == 'Drug possession']['All inmates'])
other_all = np.array(drug_data[drug_data['Offense'] == 'Other (Includes trafficking, possession, and other drug offenses.)']['All inmates'])


# Convert the 'Date' column to datetime format
dates = pd.to_datetime(drug_data['Date'], errors='coerce')


# Group data by year and sum the values for each offense
grouped_data = drug_data.groupby([dates.dt.year, 'Offense'])['All inmates'].sum().unstack()

# Plotting
width = 0.2  # Width of the bars
years = grouped_data.index
bar_positions = np.arange(len(years))

fig, ax = plt.subplots(figsize=(10, 6))

bar1 = ax.bar(bar_positions - width, grouped_data['Drug'], width, label='Drug')
bar2 = ax.bar(bar_positions, grouped_data['Drug possession'], width, label='Drug possession')
bar3 = ax.bar(bar_positions + width, grouped_data['Other (Includes trafficking, possession, and other drug offenses.)'], width, label='Other')

# Customize the plot
ax.set_xlabel('Year')
ax.set_ylabel('Number of Inmates')
ax.set_title('Inmates by Drug Offense Over Time')
ax.set_xticks(bar_positions)
ax.set_xticklabels(years)
ax.legend()

# Display the plot
plt.show()
```

### Crime Types

```python
federal_data = drug_data[drug_data['Federal or State'] == 'Federal']
state_data = drug_data[drug_data['Federal or State'] == 'State']

# Convert the 'Date' column to datetime format
dates = pd.to_datetime(drug_data['Date'], errors='coerce')

unique_years = dates.dt.year.unique()

fig, ax = plt.subplots(figsize=(14, 8))

# Grouping the data by year, federal/state, and offense, and summing the values for each year
grouped_data_federal = federal_data.groupby([dates.dt.year, 'Offense'])['All inmates'].sum().unstack()
grouped_data_state = state_data.groupby([dates.dt.year, 'Offense'])['All inmates'].sum().unstack()

# Check available columns in grouped_data_federal and grouped_data_state
print("Columns in grouped_data_federal:", grouped_data_federal.columns)
print("Columns in grouped_data_state:", grouped_data_state.columns)

# Plotting federal inmates
for offense in grouped_data_federal.columns:
    plt.bar(grouped_data_federal.index, grouped_data_federal[offense], width, label=f'Federal - {offense}')

# Plotting state inmates
for offense in grouped_data_state.columns:
    plt.bar(grouped_data_state.index, grouped_data_state[offense], width, label=f'State - {offense}', bottom=grouped_data_state.iloc[:, :grouped_data_state.columns.get_loc(offense)].sum(axis=1))

plt.xlabel('Year')
plt.ylabel('Inmates')
plt.title('Federal vs State Inmates with Specific Offenses Over Time')
plt.legend(fontsize='small')

# Display the plot
plt.show()
```
